```python
import matplotlib.pyplot as plt

# Generate 100,000 samples
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")

plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Define range
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot 3D surface
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Plot of Z = X² + Y²')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Sample data creation (replace this with your actual dataset if available)
data = {
    'HP': np.random.randint(40, 200, 100),
    'Attack': np.random.randint(30, 180, 100),
    'Defense': np.random.randint(35, 180, 100),
    'Sp. Atk': np.random.randint(30, 180, 100),
    'Sp. Def': np.random.randint(30, 180, 100),
    'Speed': np.random.randint(40, 160, 100)
}
df = pd.DataFrame(data)

# Pearson correlation
pearson_corr = df.corr(method='pearson')
print("Pearson Correlation Matrix:")
print(pearson_corr)

# Spearman correlation
spearman_corr = df.corr(method='spearman')
print("\nSpearman Correlation Matrix:")
print(spearman_corr)

# Heatmaps
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
